```python
df = pd.read_csv('cleaned_bus.csv.csv')
df.info()
```

```python
df_new = df[
    [
        "Travel Duration",
        "price",
        "Source",
        "Destination",
        "distance"
    ]
]

# Remove null values
df_new = df_new.dropna()

# Keep rows where Source OR Destination has Delhi/Bangalore
df_new = df_new[
    (
        df_new["Source"].str.contains("Delhi|Bengaluru", case=False, na=False)
    )
    |
    (
        df_new["Destination"].str.contains("Delhi|Bengaluru", case=False, na=False)
    )
]

# Save dataset
df_new.to_csv("intercitybuses.csv", index=False)
```
